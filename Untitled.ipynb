```python
from fastai.vision.all import *
from fastbook import *
from fastai.text.all import *
from fastai.vision.widgets import *
```

```python
path = Path('images')
```

```python
dblock = DataBlock()
```

```python
fns = get_image_files(path)
```

```python
def parent_labeler(o, **kwargs):
    return [Path(o).parent.name]
```

```python
animal = DataBlock(
        blocks = (ImageBlock, MultiCategoryBlock),
        get_items = get_image_files, 
        splitter = RandomSplitter(valid_pct = 0.2, seed = 42),
        get_y = parent_labeler,
        item_tfms = Resize(128))
```

```python
dls = animal.dataloaders(path)
```

```python
dls = animal.dataloaders(path)
dls.show_batch(nrows = 1, ncols = 3)
```

```python
def accuracy_multi(inp, targ, thresh = 0.5, sigmoid = True):
    if sigmoid:
        inp = inp.sigmoid()
    return ((inp > thresh) == targ.bool()).float().mean()
```

```python
learn = vision_learner(dls, resnet50,
                       metrics = partial(accuracy_multi, thresh = 0.2))
learn.fine_tune(3, base_lr = 3e-3, freeze_epochs = 4)
```

```python
learn.metrics = partial(accuracy_multi, thresh = 0.1)
learn.validate()
```

```python
learn.metrics = partial(accuracy_multi, thresh = 0.99)
learn.validate()
```

```python
preds, targs = learn.get_preds()
```

```python
accuracy_multi(preds, targs, thresh = 0.9, sigmoid = False)
```

```python
xs = torch.linspace(0.05, 0.95, 29)
accs = [accuracy_multi(preds, targs, thresh = x, sigmoid = False) for x in xs]
plt.plot(xs, accs);
```

```python
np.array(accs).max()
```

```python
xs[np.array(accs).argmax()]
```

```python
learn.metrics = partial(accuracy_multi, thresh = 0.8214)
learn.validate()
```

```python
learn = vision_learner(dls, resnet50, metrics = partial(accuracy_multi, thresh = 0.8214))
learn.fine_tune(3, base_lr = 3e-3, freeze_epochs = 4)
```

```python
learn.export()
```

```python
from fastai.vision.widgets import *
```

```python
path = Path()
upload_button = widgets.FileUpload()
output = widgets.Output()
pred_label = widgets.Label()
```

```python
learn_inf = load_learner(path/'export.pkl')
```

```python
%load_ext autoreload
%autoreload 2
```

```python
def on_data_change(change):
    pred_label.value = ''
    output.clear_output()
    im = PILImage.create(upload_button.data[-1])
    with output:
        display(im.to_thumb(256, 256))
    pred, pred_idx, probs = learn_inf.predict(im)
    if pred:
        probs = probs[pred_idx].tolist()
        probs = [str(round(prob, 5)) for prob in probs]
        pred_label.value = f"Prediction: {'; '.join(pred)}; Probability: {'; '.join(probs)}"
    else:
        pred_label.value = "Nothing is detected!"
```

```python
upload_button.observe(on_data_change, names = ['data'])
```

```python
display(VBox([widgets.Label("Select an image and surprise!"),
              upload_button,
              output,
              pred_label]))
```
